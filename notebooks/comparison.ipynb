# Model Comparison Analysis

This notebook compares the performance of different PReNet variants on Rain100H and Rain1400 datasets.

## Metrics Analyzed:
- **Loss**: Training loss value (lower is better)
- **PSNR**: Peak Signal-to-Noise Ratio in dB (higher is better)
- **SSIM**: Structural Similarity Index 0-1 (higher is better)

## Datasets:
- **Rain1400**: 1400 rainy images for training
- **Rain100H**: 100 high-resolution rainy images for testing

```python
import sys
sys.path.append('..')

from scripts.visualizer_result.comparison_analysis import ComparisonAnalyzer
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Initialize Analyzer

```python
analyzer = ComparisonAnalyzer(
    results_dir='../output/results',
    output_dir='../output/comparisons'
)
print('Analyzer initialized')
```

## Load Results from JSON Files

```python
print('Loading results from output/results/')
analyzer.load_results()
print(f'\nLoaded {len(analyzer.data_1400)} Rain1400 models')
print(f'Loaded {len(analyzer.data_100h)} Rain100H models')
```

## Run Complete Analysis

This will generate all visualizations and summary tables for both datasets

```python
analyzer.run_analysis()
```

## Rain1400 Analysis

Results saved to: `output/comparisons/rain1400/`

```python
if analyzer.data_1400:
    print('\n' + '='*80)
    print('RAIN1400 ANALYSIS')
    print('='*80)
    print('\nGenerating Rain1400 visualizations...')
    analyzer.plot_loss_comparison('Rain1400', analyzer.data_1400, 'rain1400')
    analyzer.plot_psnr_comparison('Rain1400', analyzer.data_1400, 'rain1400')
    analyzer.plot_ssim_comparison('Rain1400', analyzer.data_1400, 'rain1400')
    analyzer.plot_combined_metrics('Rain1400', analyzer.data_1400, 'rain1400')
    analyzer.generate_summary_table('Rain1400', analyzer.data_1400, 'rain1400')
else:
    print('No Rain1400 data loaded')
```

## Rain100H Analysis

Results saved to: `output/comparisons/rain100h/`

```python
if analyzer.data_100h:
    print('\n' + '='*80)
    print('RAIN100H ANALYSIS')
    print('='*80)
    print('\nGenerating Rain100H visualizations...')
    analyzer.plot_loss_comparison('Rain100H', analyzer.data_100h, 'rain100h')
    analyzer.plot_psnr_comparison('Rain100H', analyzer.data_100h, 'rain100h')
    analyzer.plot_ssim_comparison('Rain100H', analyzer.data_100h, 'rain100h')
    analyzer.plot_combined_metrics('Rain100H', analyzer.data_100h, 'rain100h')
    analyzer.generate_summary_table('Rain100H', analyzer.data_100h, 'rain100h')
else:
    print('No Rain100H data loaded')
```
